# Categorical encoding

```python
from sklearn.preprocessing import LabelEncoder

lb = LabelEncoder()
y['Diagnosis'] = lb.fit_transform(y[['Diagnosis']])
```

# Train the model and fix imbalance

```python
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
smote = SMOTE()
X_train, y_train = smote.fit_resample(X_train, y_train)
```

# Fix outliers

```python
def fix_outliers_1(feature):
    train_iqr = X_train[feature].quantile(0.75) - X_train[feature].quantile(0.25)
    train_lower_bound = X_train[feature].quantile(0.25) - (train_iqr*1.5)
    train_upper_bound = X_train[feature].quantile(0.75) + (train_iqr*1.5)
    X_train[feature] = np.clip(X_train[feature], train_lower_bound, train_upper_bound)


for feature in X_train.columns:
    fix_outliers_1(feature)



def fix_outliers_2(feature):
    test_iqr = X_test[feature].quantile(0.75) - X_test[feature].quantile(0.25)
    test_lower_bound = X_test[feature].quantile(0.25) - (test_iqr*1.5)
    test_upper_bound = X_test[feature].quantile(0.75) + (test_iqr*1.5)
    X_test[feature] = np.clip(X_test[feature], test_lower_bound, test_upper_bound)


for feature in X_test.columns:
    fix_outliers_2(feature)
```

# Scale the skewd features

```python
dataset_skew_features = X_train.skew()
dataset_skew_features_2 = X_test.skew()
transformed_features = []
boxcox_lambdas = {}

from scipy.stats import boxcox

# Train data
for feature in X_train.columns:
    skewness = dataset_skew_features[feature]

    
    if skewness>1:
        if (X_train[feature]>0).all():
            X_train[feature] = np.log1p(X_train[feature])
        else:
            X_train[feature] = np.sqrt(X_train[feature] - X_train[feature].min()+1)
        transformed_features.append(feature)
        
    elif skewness<-1:
        X_train[feature], boxcox_lambda = boxcox(X_train[feature] - X_train[feature].min() + 1)
        boxcox_lambdas[feature] = boxcox_lambda
        transformed_features.append(feature)

# Test data
for feature in X_test.columns:
    skewness = dataset_skew_features_2[feature]


    if skewness>1:
        if(X_test[feature]>0).all():
            X_test[feature] = np.log1p(X_test[feature])
        else:
            X_test[feature] = np.sqrt(X_test[feature] - X_test[feature].min() + 1)

    elif skewness<-1 and feature in boxcox_lambdas:
        X_test[feature] = boxcox(X_test[feature] - X_test[feature].min() + 1, boxcox_lambdas[feature])
```

# Apply standard scaler on less skewed features.

```python
from sklearn.preprocessing import StandardScaler

less_skewed = [feature for feature in X_train.columns if feature not in transformed_features]

scale = StandardScaler()
X_train[less_skewed] = scale.fit_transform(X_train[less_skewed])
X_test[less_skewed] = scale.transform(X_test[less_skewed])
```

# Feature selection

```python
def select_feature(dataset, threshod):
    corr_set = set()
    corr_matrix = dataset.corr()

    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if np.abs(corr_matrix.iloc[i, j])>threshod:
                col_name = corr_matrix.columns[i]
                corr_set.add(col_name)
    return corr_set

del_features = select_feature(X_train, 0.7)
X_train.drop(del_features, axis=1, inplace=True)
X_test.drop(del_features, axis=1, inplace=True)
```

# Save the datasets

```python
X_train.to_csv('X_train.csv', index=False)
X_test.to_csv('X_test.csv', index=False)
y_train.to_csv('y_train.csv', index=False)
y_test.to_csv('y_test.csv', index=False)
```
